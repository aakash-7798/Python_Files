```python
## We add a column which consists of Phone numbers so i use random package for random number generating
## Below is an Example
rd.randrange(9000000000,9999999999,10)
```

```python
## Now i want 30 records so i create a list 
Ph = []
for i in range(0,30):
    Ph.append(rd.randrange(9000000000,9999999999,10))
```

```python
## Adding New Column as Phone Number 
df[["Phone Number"]] = pd.DataFrame(Ph)
```

```python
## If we observe there are few values which are I but actually they are 1 so lets clean them
for i in range(0,30):
    df["Roll Number"][i] = df["Roll Number"][i].replace(" I","1")
```

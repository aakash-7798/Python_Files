```python
def findMinDiff(arr, n, m):
    if (m==0 or n==0):
        return 0
    arr.sort()
    print(arr)
    if (n < m):
        return -1
    min_diff = arr[n-1] - arr[0]
    print("Initially Min diff : ",min_diff)
    for i in range(len(arr) - m + 1):
        min_diff = min(min_diff , arr[i + m - 1] - arr[i])
        print(f" arr[{i + m - 1}] = {arr[i + m - 1]}, arr[{i}] = {arr[i]}  and  {arr[i + m - 1]} - {arr[i]}= {arr[i + m - 1] - arr[i]}")
    return min_diff
arr = [12, 4, 7, 9, 2, 23, 25, 41,30, 40, 28, 42, 30, 44, 48,43,50]
print("Minimum difference is", findMinDiff(arr,len(arr),7))
```

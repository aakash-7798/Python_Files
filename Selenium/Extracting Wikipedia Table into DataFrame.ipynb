```python
from selenium import webdriver
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
```

```python
driver = webdriver.Chrome("C:\\WebDriver\\chromedriver.exe")
driver.get(url)
```

```python
# --- get table ---

headers = []
columns = dict()
table_id = driver.find_element_by_id('constituents')
all_rows = table_id.find_elements_by_tag_name("tr")
```

```python
table_id
```

```python
all_rows[0].text
```

```python
all_rows[0].find_elements_by_tag_name("th")
```

```python
for i in range(len(all_rows[0].find_elements_by_tag_name("th"))):
    print(all_rows[0].find_elements_by_tag_name("th")[i].text)
```

```python
# --- headers ---

row = all_rows[0]
all_items = row.find_elements_by_tag_name("th")
for item in all_items:
    name = item.text
    columns[name] = []
    headers.append(name)

print(headers)
```

```python
headers
```

```python
columns
```

```python
all_rows[1].find_element_by_tag_name('td')
```

```python
for i in all_rows[1:]:
    print(i.find_element_by_tag_name('td').text)
```

```python
# --- data ---

for row in all_rows[1:]:
    all_items = row.find_elements_by_tag_name("td")
    for name, item in zip(headers, all_items):
        value = item.text
        columns[name].append(value)
```

```python
columns
```

```python
df = pd.DataFrame(columns)
df
```

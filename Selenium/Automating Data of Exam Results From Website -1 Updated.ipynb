```python
url = 'http://www.results.manabadi.co.in/osmania-university-ug-cbcs-ba-bcom-bsc-bba-2019-exam-results-13-082019.htm'
# ht = input("Enter Hall Ticket Number : ")
# cs = input("Enter Course : Bcom Bsc BA BBA ")
driver = webdriver.Chrome(chrome_options=options,executable_path="C:\\WebDriver\\chromedriver.exe")
driver.get(url)
# ht = ["105916467074","105916487050","105916467084","105916467086","105916467057"]
ht = []
for i in range(70,75):
    ht.append(f'105916467{i:03}')
Check_Result = []
for i in range(len(ht)):
    driver.find_element_by_id("htno").send_keys(ht[i])
    time.sleep(1)
    driver.find_element_by_id("Degree").send_keys("BSC")
    time.sleep(1)
    driver.find_element_by_id('btnsubmit').click()
    time.sleep(1)
    driver.find_element_by_id("htno").clear()
    if [driver.find_element_by_xpath('//span[@id="sMsg"]').text][0].strip(".") == 'Invalid Hall Ticket No':
        Check_Result.append('Invalid Hall Ticket No')
        continue
    else:
        Table1 = driver.find_element_by_xpath('//div[@id="subhtm"]//table[1]')
        Table2 =driver.find_element_by_xpath('//div[@id="subhtm"]//table[2]')
        Table3 = driver.find_element_by_xpath('//div[@id="subhtm"]//table[3]')
        Personal_Information = []
        for i in Table1.find_elements_by_tag_name('td'):
            Personal_Information.append(i.text)
        Personal_Information = Personal_Information[1:]
        PI = {}
        for (i,j) in zip(Personal_Information[::2],Personal_Information[1::2]):
            PI[i] = [j]
        Marks_Details = []
        for i in Table2.find_elements_by_tag_name('td'):
            Marks_Details.append(i.text)
        Marks_Details = Marks_Details[1:]
        Marks_Details[0:5]
        # [Marks_Details[5*i:5*(i+1)] for i in range(int(len(Marks_Details[5:])/10) + 1)]
        MI = [Marks_Details[i::5][1:] for i in range(len(Marks_Details[0:5])) ]
        MD = {}
        for (i,j) in zip(Marks_Details[0:5],MI):
            MD[i] = j
        Result = []
        for i in Table3.find_elements_by_tag_name('td'):
            Result.append(i.text)
        Result = Result[1:]
        RT = pd.DataFrame(columns=Result[:3])
        lst = [list(i) for i in list(np.array_split(Result[3:],int(len(Result[3:])/3)))]
        for i in range(len(lst)):
            RT.loc[i] = lst[i]
        RT = RT.to_dict(orient='list')
        Check_Result.append(list(zip([PI],[MD],[RT])))
```

```python
roll= input("Enter Last Three Digits :")
print("   ")
for i in ht:
    if i[9:] == roll:
        if Check_Result[ht.index(i)] == 'Invalid Hall Ticket No':
            print("-" * 30)
            print("   ",'Invalid Hall Ticket No')
        else:     
            print(tabulate(pd.DataFrame(Check_Result[ht.index(i)][0][0]).transpose(),tablefmt='pretty'))
            print(" ")
            print(tabulate(pd.DataFrame(Check_Result[ht.index(i)][0][1]),showindex=False,headers =pd.DataFrame(Check_Result[0][0][1]).columns,tablefmt='pretty'))
            print(" ")
            print(tabulate(pd.DataFrame(Check_Result[ht.index(i)][0][2]),showindex=False,headers =pd.DataFrame(Check_Result[0][0][2]).columns,tablefmt='pretty'))
```

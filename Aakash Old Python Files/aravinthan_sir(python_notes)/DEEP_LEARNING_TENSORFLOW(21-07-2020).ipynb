```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = pd.read_csv("Car_Purchasing_Data.csv",encoding="ISO-8859-1")
df = pd.read_csv("Car_Purchasing_Data.csv")
```

```python
df
```

```python
df.columns
```

```python
X = df[['Gender', 'Age',
       'Annual Salary', 'Credit Card Debt', 'Net Worth']]
```

```python
y = df["Car Purchase Amount"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_scaled = MinMaxScaler().fit_transform(X)
```

```python
X_scaled
```

```python
y = y.values.reshape(-1,1)
```

```python
y
```

```python
y_scaled = MinMaxScaler().fit_transform(y)
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y_scaled)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow.keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
### Input Layer
model.add(Dense(30,input_dim = 5,activation = 'relu'))
```

```python
### Hidden Layer
model.add(Dense(30,activation= 'relu'))
```

```python
### Final Layer
model.add(Dense(1,activation = 'linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss = "mean_squared_error")
```

```python
epochs_hist = model.fit(x_train,y_train,epochs=30,batch_size=50,verbose=1,validation_split=0.3)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

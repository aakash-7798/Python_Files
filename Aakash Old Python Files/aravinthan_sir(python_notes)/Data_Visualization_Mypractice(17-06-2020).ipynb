# VARIETIES OF SEABORN SCATTER

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
iris = pd.read_csv('Iris.csv')
tips = pd.read_csv('tips.csv')
```

```python
iris.head(10)
```

```python
tips.head(10)
```

```python
fig,ax = plt.subplots(4,3)
fig.set_size_inches(30,25)
plt.rcParams["axes.labelsize"] = 10
sns.set(font_scale = 1)
sns.scatterplot(x = "total_bill",y = "tip",data=tips,ax=ax[0,0])
sns.scatterplot(x="total_bill", y="tip", hue="time",data=tips,ax=ax[0,1])
sns.scatterplot(x="total_bill", y="tip",hue="time", style="time", data=tips,ax=ax[0,2])
sns.scatterplot(x="total_bill", y="tip",hue="day", style="time", data=tips,ax=ax[1,0])
sns.scatterplot(x="total_bill", y="tip", size="size",data=tips,ax=ax[1,1])
sns.scatterplot(x="total_bill", y="tip",hue="size", size="size",data=tips,ax=ax[1,2])
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
sns.scatterplot(x="total_bill", y="tip",hue="size", size="size",sizes=(20, 200), palette=cmap,legend="full", data=tips,ax=ax[2,0])
sns.scatterplot(x="total_bill", y="tip",hue="size", size="size",sizes=(20, 200), palette=cmap,legend="full", data=tips,ax=ax[2,1])
sns.scatterplot(x="total_bill", y="tip",hue="day", size="smoker",palette="Set2",data=tips,ax=ax[2,2])
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(x="total_bill", y="tip", style="time",markers=markers,data=tips,color = "black",ax=ax[3,0])
sns.scatterplot(x="total_bill", y="tip",s=100, color=".2", marker="+",data=tips,ax=ax[3,1])
# sns.relplot(x="total_bill", y="tip",col="time", hue="day", style="day",kind="scatter", data=tips,ax=ax[3,2])
```

# RANK

```python
df = pd.read_csv("nba.csv").dropna(how = "all")
```

```python
df.columns
df
```

```python
df.isna().sum()
```

```python
df['Salary'] = df["Salary"].fillna(0).astype("int")
```

```python
df.info()
```

```python
df["Salary"].rank()
```

```python
df["Salary rank"] = df["Salary"].rank(ascending = False).astype(int)
df
```

```python
df.sort_values(["Salary rank"],ascending = True)
```

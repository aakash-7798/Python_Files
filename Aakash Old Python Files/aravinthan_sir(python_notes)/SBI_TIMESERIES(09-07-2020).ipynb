```python
import pandas as pd
```

```python
df = pd.read_csv('SBIN.NS (3).csv')
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
import plotly.express as px


import plotly.graph_objects as go
```

```python
fig = go.Figure([go.Scatter(x=df['Date'], y=df['Open'])])
fig.show()
```

```python
def add_features(df):
    df['year'] = df['Date'].dt.year
    df['month'] = df['Date'].dt.month
    df['day'] = df['Date'].dt.day
    df['dayofweek'] = df['Date'].dt.dayofweek
```

```python
add_features(df)
```

```python
df
```

```python
fig = go.Figure([go.Scatter(x=df['year'], y=df['Open'])])
fig.show()
```

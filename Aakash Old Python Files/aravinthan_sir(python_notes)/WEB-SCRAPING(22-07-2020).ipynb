```python
from bs4 import BeautifulSoup
import requests
```

```python
# url = "https://plotly.com"
url = "https://boston.craigslist.org/search/sof"
```

```python
response = requests.get(url)
```

```python
response
```

```python
df = response.text
```

```python
soup = BeautifulSoup(df,'html.parser')
```

```python
soup.find_all('p',{'class':'result-info'})
```

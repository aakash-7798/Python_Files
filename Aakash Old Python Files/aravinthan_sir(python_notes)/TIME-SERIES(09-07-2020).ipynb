```python
import pandas as pd
```

```python
df = pd.read_csv("SBIN.NS (3).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
def add_features(df):
    df['year'] = df['Date'].dt.year
    df['month'] = df['Date'].dt.month
    df['day'] = df['Date'].dt.day
    
add_features(df)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df,x = 'Date',y = 'Open')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_size_inches(15,5)
sns.lineplot(x="Date",y="Open",data = df[df['year'] == 2019],ax=ax1)
sns.lineplot(x="Date",y="Open",data = df[df['year'] == 2018],ax=ax2)
ax1.title.set_text("SBI_2019 Trend")
ax2.title.set_text("SBI_2018 Trend" )
```

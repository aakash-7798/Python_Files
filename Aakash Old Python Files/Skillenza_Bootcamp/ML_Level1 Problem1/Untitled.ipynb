```python
corpus = ['This is an excellent movie','The movie was fantastic I liked it',
          'horrible acting','waste of money']
```

```python
from nltk.tokenize import word_tokenize
all_words =[]
for i in corpus:
    tokenize_word = word_tokenize(i)
    for j in tokenize_word:
        all_words.append(j)
      
```

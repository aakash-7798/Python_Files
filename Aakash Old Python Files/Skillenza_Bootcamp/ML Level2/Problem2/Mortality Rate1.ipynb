```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
lr = LinearRegression()
le = LabelEncoder()
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
NC = ['Fever (temperature ≥37·3°C)',
       'Cough', 'Diabetes', 'Hypertension',
       'Cardiovascular Disease (incl. CAD)',
       'Lymphocyte Count (10^9/L) - Median',
       'White Blood Cell Count (10^9/L) - Median',
       'Chronic obstructive lung (COPD)', 'Diarrhoea',
       'White Blood Cell Count (10^9/L) - LQ', 'Mean Age',
       'White Blood Cell Count (10^9/L) - UQ', 'Chronic kidney/renal disease',
       'Shortness of Breath (dyspnoea)', 'Lymphocyte Count (10^9/L) - UQ',
       'Lymphocyte Count (10^9/L) - LQ', 'Cancer (Any)']
```

```python
df["Severity_label"]= pd.DataFrame(le.fit_transform(df["Severity"]))
df["Severity_scaler"] = scaler.fit_transform(df[["Severity_label"]])
```

```python
df = df.reindex(columns = ['ID', 'Mortality', 'Severity','Severity_label',
       'Severity_scaler', '% Male', 'Fever (temperature ≥37·3°C)',
       'Cough', 'Diabetes', 'Hypertension',
       'Cardiovascular Disease (incl. CAD)',
       'Lymphocyte Count (10^9/L) - Median',
       'White Blood Cell Count (10^9/L) - Median',
       'Chronic obstructive lung (COPD)', 'Diarrhoea',
       'White Blood Cell Count (10^9/L) - LQ', 'Mean Age',
       'White Blood Cell Count (10^9/L) - UQ', 'Chronic kidney/renal disease',
       'Shortness of Breath (dyspnoea)', 'Lymphocyte Count (10^9/L) - UQ',
       'Lymphocyte Count (10^9/L) - LQ', 'Cancer (Any)'])
```

```python
df1 = df.copy()
df1 = df1.drop(["Severity","Severity_label"],axis=1)
```

```python
new_values = []
for i in range(len(NC)):
    lr.fit(df1[df1[NC[i]].notnull()].iloc[:,[0,1,2,3,4+i]].iloc[:,0:4],df1[df1[NC[i]].notnull()].iloc[:,[0,1,2,3,4+i]][[NC[i]]])
    new_values.append(lr.predict(df1[df1[NC[i]].isnull()].iloc[:,[0,1,2,3,4]].iloc[:,0:4]))
```

```python
notnull = []
isnull = []
for  i in range(len(NC)):
    notnull.append(df1[df1[NC[i]].notnull()].iloc[:,[0,1,2,3,4+i]])
    isnull.append(df1[df1[NC[i]].isnull()].iloc[:,[0,1,2,3,4+i]])
    notnull[i].reset_index(drop=True,inplace = True)
    isnull[i].reset_index(drop=True,inplace = True)    
    isnull[i][NC[i]]  = pd.DataFrame(new_values[i])[0] 
    df1.iloc[:,4+i] = pd.concat([notnull[i],isnull[i]],ignore_index=True).iloc[:,-1]
```

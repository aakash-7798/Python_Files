```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Data_and_Train_Data/train_data.xlsx")
```

```python
df.head()
```

```python
# pd.options.display.float_format = '{:.2f}'.format
df.Holiday = df.Holiday.replace({False:0,True:1})
df["month"] = df.Week_Date.dt.month
df
```

```python
df = df.reindex(columns = ['Book_Stall', 'Genre', 'Week_Date','month', 'Holiday','Sales_in_a_week'])
df
```

```python
df = df.drop(["Week_Date"],axis=1)
df
```

```python
test = pd.read_excel("test_sales.xlsx")
test
```

```python
test.columns
```

```python
# pd.options.display.float_format = '{:.2f}'.format
test.Holiday = test.Holiday.replace({False:0,True:1})
test["month"] = test.Week_Date.dt.month
test = test.reindex(columns = ['ID', 'Book_Stall', 'Genre', 'Week_Date','month', 'Holiday'])
```

```python
test = test.drop("Week_Date",axis=1)
test
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor 
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
```

```python
dt.fit(df.drop("Sales_in_a_week",axis=1),df.Sales_in_a_week)
rf.fit(df.drop("Sales_in_a_week",axis=1),df.Sales_in_a_week)
```

```python
predict_dt = pd.DataFrame(dt.predict(test.drop("ID",axis=1)))
predict_dt
```

```python
predict_rf = pd.DataFrame(rf.predict(test.drop("ID",axis=1)))
predict_rf
```

```python
Sales_dt =pd.read_csv("Submission_format_Sales.csv")
Sales_dt.Sales_in_a_week = predict_dt[0]
Sales_dt
```

```python
Sales_rf =pd.read_csv("Submission_format_Sales.csv")
Sales_rf.Sales_in_a_week = predict_rf[0]
Sales_rf
```

```python
text = []
for i in range(len(df)):
    text.append(re.findall(r'\d+',df["text"][i]))
```

```python
import ast
text_new = []
for i in range(len(df)):
    text_new.append([ast.literal_eval(j) for j in text[i]]) 
```

```python
l = []
for i in range(len(text_new)):
    l.append(len(text_new[i]))   
```

```python
pd.set_option("display.max_rows",None)
pd.DataFrame(pd.DataFrame(l)[0].value_counts())
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
text_new
```

```python
voc_size = 10000
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sent_length=25
embedded_docs=pad_sequences(text_new,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedding_vector_features=25
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
len(embedded_docs),y.shape
```

```python

import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

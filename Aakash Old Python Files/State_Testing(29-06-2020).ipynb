```python
import pandas as pd
```

```python
st = pd.read_csv("D:/PythonFiles/Covid_India_Datasets/StatewiseTestingDetails.csv")
```

```python
st.drop(columns=["Latitude","Longitude"],inplace=True)
```

```python
st
```

```python
loc =  st.State.value_counts().index[:].to_list()
```

```python
loc[0]
```

```python
import geocoder
lt = []
lg = []
for i in range(len(loc)):
    lt.append(geocoder.osm(loc[i]).json["lat"])
    lg.append(geocoder.osm(loc[i]).json["lng"])
```

```python
# lt
```

```python
# lg
```

```python
# direc = pd.concat([pd.Series(lt),pd.Series(lg)],axis=1)
# direc.columns = ["Lat","Lng"]
# direc
```

```python
st["Lat"] = st.State.replace(loc,lt)
st["Lng"] = st.State.replace(loc,lg)
```

```python
st.head()
```

```python
st.isna().sum()
```

```python
st.Positive.fillna(st.Positive.median(),inplace = True)
```

```python
st.Negative.fillna(st.Negative.median(),inplace = True)
```

```python
st.isna().sum()
```

```python
import plotly.express as px
```

```python
fig = px.density_mapbox(st,lat="Lat",lon="Lng",height=700,zoom=3.55,radius=7,
                        animation_frame="Date",hover_name="State",
                        color_continuous_scale="Portland",hover_data=["TotalSamples","Negative","Positive"])
fig.update_layout(mapbox_style = "open-street-map")
fig.update_layout(margin =  {'r':0,'t':0,'l':0,"b":0})
fig.show()
```

```python
import nltk
import os
import nltk.corpus
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.data import load
nltk.download('averaged_perceptron_tagger')
nltk.download('gutenberg')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
halmet_pos = []
for i in halmet[:2000]:
    word_pos = nltk.pos_tag([i])
    halmet_pos.append(word_pos)
```

```python
halmet_nnp = []
for i in halmet_pos:
    if i[0][1]=="NNP":
        halmet_nnp.append(i[0][0])
```

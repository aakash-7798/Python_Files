```python
x = df[['deposit_amount_2011',
       'deposit_amount_2012', 'deposit_amount_2013', 'deposit_amount_2014',
       'deposit_amount_2015', 'deposit_amount_2016']]
y = df[['deposit_amount_2017']]
```

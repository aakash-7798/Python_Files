LOADING DATA DIRECTLY FROM YAHOO FINANCE WEBSITE WITH THE THE HELP OF PANDAS DATAREADER
I'M TAKING DATA FROM 2010 TO TILL NOW(i.e 2020)

```python
#### Following variable stocks has companies names with their stocked names
stocks = [ "AXISBANK.NS","BHARTIARTL.NS","HDFCBANK.NS","HDFC.NS","TCS.NS","HCLTECH.NS","INFY.NS","KOTAKBANK.NS","MARUTI.NS","RELIANCE.NS"]
data =  web.DataReader(stocks,"yahoo",dt.datetime(2010,1,1),dt.datetime.now())
```

AS WE HAVE ONLY TEN COMPANIES SO I WILL CREATE 10 DATAFRAMES AND KEEPING EACH COMPANY DETAILS IN EACH DATAFRAME AFTER THAT I WILL ADD ALL THE DATAFRAMES INTO SINGLE DATAFRAME

```python
# With the help of for loop i will create as many dataframes as i want
y = []                                  # y is list where all dataframes are stored
for i in range(len(stocks)):            # stocks has 10 companies so range starts from 0 to 10 
               y.append(pd.DataFrame({},columns = ["Date","Company","Open","High","Low","Close","Adjclose","Volume"]))
```

```python
### The Following is the First Dataframe 
### With indexing we can achieve that
y[0]
### Similarly for all we can see all dataframes   
```

LOADIND DATA INTO EACH DATAFRAME WITH RESPECT TO THIER COMPANIES

```python
for i in range(len(stocks)):
        y[i][y[i].columns[0]] = df["Date"]
        y[i][y[i].columns[1]] = pd.DataFrame(stocks)[0].str.split(".")[i][0]
        y[i][y[i].columns[2]] = df["Open"][stocks[i]]
        y[i][y[i].columns[3]] = df["High"][stocks[i]]
        y[i][y[i].columns[4]] = df["Low"][stocks[i]]
        y[i][y[i].columns[5]] = df["Close"][stocks[i]]
        y[i][y[i].columns[6]] = df["Adj Close"][stocks[i]]
        y[i][y[i].columns[7]] = df["Volume"][stocks[i]]
```

```python
## Appending all Dataframes
df = y[0].append([y[1],y[2],y[3],y[4],y[5],y[6],y[7],y[8],y[9]])
```

## I AM CONSIDERING PROPHET FOR PREDICTIONS

IF WE CAN REMEMBER AT STARTING I HAVE CREATED DATAFRAMES ACCORDING TO COMPANY WISE WHERE ALL WERE STORED IN A LIST Y

df.Company.unique() GIVES ME AS ['AXISBANK', 'BHARTIARTL', 'HDFCBANK', 'HDFC', 'TCS', 'HCLTECH','INFY', 'KOTAKBANK', 'MARUTI', 'RELIANCE']

y[0] REPRESENTS ALL DETAILS OF AXISBANK AND SIMILARLY FOR ALL

FOR PROPHET WE NEED ONLY TWO COLUMNS WHERE ONE IS DATE AND ANOTHER IS OUR TARGET

COLUMN NAMES SHOULD AS ALSO BE AS ds and y

AS I HAVE STORED ALL DATA IN SINGLE DATAFRAME WHERE WE CANNOT MAKE PREDICTIONS FOR PARTICULAR COMPANY SO I'M RECALLING THOSE DATAFRAMES WHERE I HAVES STORED IN Y AND MAKING A FORLOOP TO EXTRACT DATE AND CLOSE TO KEEP THEM INTO NEW DATAFRAME WITH COLUMNS CHANGED AND ALSO SPLITTIG DATA  INTO TRAIN AND TEST 

```python
y_new = []
for i in range(len(stocks)):
  y_new.append(pd.DataFrame())                    #### Creating Dataframes and keeping them in y_new
  y_new[i]  = (y[i].loc[:,["Date","Close"]])      ### Extracting only date and close
  y_new[i].columns = ["ds","y"]                   #### Changing column names
### If we want train and test for verification we can go but as of now I'm considering
  # trains[i] = (y_new[i].iloc[0:int(len(y_new[i])*0.95),:])  #### This stores the 95% of  y_new as train 
  # tests[i]  = (y_new[i].iloc[int(len(y_new[i])*0.05):len(y_new[i]),:])  #### This stores the 5% of  y_new as test
```

NOW WE CALL PROPHET

```python
from fbprophet import Prophet
```

```python
fits_df ,futures,forecasts = [],[],[]
for i in range(len(stocks)):
  futures.append(pd.DataFrame())           #### Creating Dataframes and keeping them in futures
  forecasts.append(pd.DataFrame())         #### Creating Dataframes and keeping them in forecasts
  fits_df.append(pd.DataFrame())        #### Creating Dataframes and keeping them in trains_fit
  fits_df[i] = Prophet().fit(y_new[i])         #### We Want to fit each dataframe so i have created fits_df which stores the list of fitted dataframes 
  futures[i] = fits_df[i].make_future_dataframe(periods=12,freq = "MS")   ## Futures stores list of  fututre_dataframes
  forecasts[i] = fits_df[i].predict(futures[i])                          ### forecasts stores list of predicted futures dataframe
```
